```python
cat_infra=["Glass, Ceramics & Concrete","Building Materials","Civil Engineering",
       "Railroad Manufacture","Transportation/Trucking/Railroad"]
cat_machine=["Machinery","Mechanical or Industrial Engineering"]
cat_auto=["Automotive"]
cat_EE=["Electrical/Electronic Manufacturing","Semiconductors","Consumer Electronics"]
cat_biomed=["Medical Devices","Biotechnology"]
cat_chem=["Chemicals"]
cat_other=["Mining & Metals","Packaging and Containers","Paper & Forest Products","Plastics", "Printing"]
```

```python
clist_infra=[]
clist_machine=[]
clist_auto=[]
clist_EE=[]
clist_chem=[]
clist_biomed=[]
clist_other=[]
for row in ind_comp_list2.itertuples():
    if row.industry in cat_infra:
        clist_infra.append(row.dataset_id)
    elif row.industry in cat_machine:
        clist_machine.append(row.dataset_id)
    elif row.industry in cat_auto:
        clist_auto.append(row.dataset_id)
    elif row.industry in cat_EE:
        clist_EE.append(row.dataset_id)
    elif row.industry in cat_chem:
        clist_chem.append(row.dataset_id)
    elif row.industry in cat_other:
        clist_other.append(row.dataset_id)
    elif row.industry in cat_biomed:
        clist_biomed.append(row.dataset_id)

clist_manu=clist_infra+clist_machine+clist_auto+clist_EE+clist_chem+clist_biomed+clist_other
```
